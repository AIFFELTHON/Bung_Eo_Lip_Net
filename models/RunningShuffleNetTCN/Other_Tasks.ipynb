## korean_detected_face.csv 작성, avi 파일명 변경, 영상 align txt 작성

```python
import os
import re


word_list = ['특히', '하지만', '그리고', '이렇게', '함께', '어떤', '이번', '오늘', '다른', '코로나', '확진자', '뉴스에서', '미국', '한국', '한국어']
word_dict = {
    'especially': '특히',
    'but': '하지만',
    'and': '그리고',
    'eruke': '이렇게',
    'with': '함께',
    'othun': '어떤',
    'this': '이번',
    'today': '오늘',
    'differ': '다른',
    'covid': '코로나',
    'huak': '확진자',
    'atnews': '뉴스에서',
    'usa': '미국',
    'korea': '한국',
    'korean': '한국어'
}
dir_list = ['train', 'val', 'test']
data_dir = '/home/PHR/Lipreading_using_TCN_running/hangeul'

# for word in word_list:
#     for dir in dir_list:
#         dir_path = f'{data_dir}/{word}/{dir}' 
#         if not os.path.exists(dir_path):                      
#             os.makedirs(dir_path)  # 디렉토리 생성
#             print(f'{word}/{dir} 폴더 생성')


csv_path = f'{data_dir}/korean_detected_face.csv'
f = open(csv_path, 'w')


for word in word_list:
    for dir in dir_list:
        dir_path = f'{data_dir}/{word}/{dir}'
        nums_list = []
        for file in os.listdir(dir_path):
            if file.endswith('.avi'):
                # --- avi 파일명 변경 ---
                file_oldname = f'{dir_path}/{file}'
                file_name, nums_ext = file.split('_')
                nums, ext = nums_ext.split('.')
                nums = nums.zfill(5)
                nums_list.append(nums)

                if not re.compile('[가-힣]+').findall(file_name):
                    file_newname = word_dict[file_name]
                file_newname = f'{dir_path}/{file_name}_{nums}'
                video_filename = f'{file_newname}.{ext}'
                os.rename(file_oldname, video_filename)

                # --- txt 파일 생성 ---
                txt_filename = f'{file_newname}.txt'
                txt_contents = 'Disk reference: 1\nChannel: Sejong\nProgram start: 2022-05-30 18:40:00 +0000\nClip start: 63.3 seconds\nDuration: 1.2 seconds'

                with open(txt_filename, 'w') as txt_file:
                    txt_file.write(txt_contents)

        nums_list.sort()
        for nums in nums_list:
            # --- CSV 작성 ---
            csv_str = f'{file_name}/{dir}/{file_name}_{nums},0\n'
            f.writelines(csv_str)



f.close()
```
