# dict.json 경로 불러오기

```python
import os
path='./video'
```

```python
file = open(path + 'dict.json', "r")
text = file.read() 
wordList = text.split()
wordCount = {} 
 
for word in wordList:
    # Get 명령어를 통해, Dictionary에 Key가 없으면 0리턴
    if word == '".':
        pass
    wordCount[word] = wordCount.get(word, 0) + 1    
    keys = sorted(wordCount.keys())
 
for word in keys:
    if wordCount[word] >= 20:
        print(word + ':' + str(wordCount[word]))
```

## 시각화 하기

```python
from matplotlib import pyplot as plt
x,y =[],[]  
for word in keys:
    if wordCount[word] >= 25 and wordCount[word] <= 100:
        print(word + ':' + str(wordCount[word]))
        y.append(word[1:])
        x.append(wordCount[word])
x.append(50)
y.append('안녕하십니까')
```

```python
# 세로 그래프
plt.figure(figsize=(3, 10))
plt.plot(x,y,
         color='black',
         marker='o', markerfacecolor='blue',
         markersize=5)

plt.show()
```

```python
# 가로 그래프
plt.figure(figsize=(20, 5))
plt.plot(y, x,
         color='skyblue',
         marker='o', markerfacecolor='blue',
         markersize=6)

plt.show()
```

# MostWord 시각화 하기

빈도수 높은 단어들을 모은 영상을 train/test/val로 나눔

```python
word = ['특히', '하지만', '그리고', '이렇게', '함께', '어떤', '이번', '오늘', '다른', '코로나', '확진자', '뉴스에서', '미국', '한국', '한국어']

train = [30, 38, 32, 18, 28, 28, 26, 26, 30, 16, 26, 38, 20, 14, 16]
val = [3, 4, 3, 3, 3, 3, 3, 3, 3, 2, 3, 3, 3, 2, 2]
test = [2, 4, 3, 2, 3, 3, 3, 2, 3, 2, 3, 2, 3, 2, 2]
```

```python
MostWord = {"word": word, "train" : train, "val": val, "test": test}
```

```python
mw = pd.DataFrame(MostWord)
mw
```

```python
w=0.15
nrow = mw.shape[0]
idx = np.arange(nrow)
idx
```

```python
mw.plot(kind='bar', figsize=(10,6), stacked=True, alpha=0.7)
ax = plt.subplot()

plt.xlabel('word')
plt.ylabel('video_num')
plt.title('MostWord', fontsize=16, fontweight= 'bold')
plt.xticks(idx, mw['word'], rotation = 30)

#ax.axis([0, 15, 0, 50])
for i, v in enumerate(mw['word']):
    ax.text(i, mw['train'][i]-10, mw['train'][i], horizontalalignment='center')
    ax.text(i, mw['train'][i]+mw['val'][i]-2, mw['val'][i], ha="center")
    ax.text(i, mw['train'][i]+mw['val'][i]+mw['test'][i]-2, mw['test'][i], ha="center")

plt.show()
```
